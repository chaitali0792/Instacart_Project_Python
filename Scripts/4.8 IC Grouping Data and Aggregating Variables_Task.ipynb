Step 6) The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Creating the new column for mean price of item for each user_id
df_ords_prods_merge['Spending'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Creating flag based on mean price ('Low spender')
df_ords_prods_merge.loc[df_ords_prods_merge['Spending'] < 10, 'Spending_Flag'] = 'Low spender'
```

```python
# 'High Spender'
df_ords_prods_merge.loc[df_ords_prods_merge['Spending'] >= 10, 'Spending_Flag'] = 'High spender'
```

```python
# checking count of 'Spending flag'
df_ords_prods_merge['Spending_Flag'].value_counts(dropna = False)
```

```python
# # Checking important columns data
df_ords_prods_merge[['user_id', 'Spending_Flag', 'Spending']].head(60)
```

Step 7) In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

# 5. Create order frequency flag for each user based on days_since_prior_order

```python
# Creating the new column for median of 'days_since_prior_order' for each user_id
df_ords_prods_merge['Frequency'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Creating flag based on median ('Non-frequent customer')
df_ords_prods_merge.loc[df_ords_prods_merge['Frequency'] > 20, 'Order_Frequency_Flag'] = 'Non-frequent customer'
```

```python
# 'Regular customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['Frequency'] > 10) & (df_ords_prods_merge['Frequency'] <= 20), 'Order_Frequency_Flag'] = 'Regular customer'
```

```python
# 'Frequent customer'
df_ords_prods_merge.loc[df_ords_prods_merge['Frequency'] <= 10, 'Order_Frequency_Flag'] = 'Frequent customer'
```

```python
# Checking the count of frequency flag
df_ords_prods_merge['Order_Frequency_Flag'].value_counts(dropna=False)
```

```python
# Checking important columns data
df_ords_prods_merge[['user_id', 'Order_Frequency_Flag', 'Frequency']].head(60)
```
